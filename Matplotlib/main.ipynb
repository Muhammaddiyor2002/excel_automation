```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
plt.plot([1,2,3,4])
```

```python
fig, ax = plt.subplots()

ax.plot([1,2,3,4])
```

```python
x = [1,2,3,4,5,6,4]
y = [2,3,2,1,6,10,3]
c = [1,1,2,2,3,4,4]

fig, ax=plt.subplots()
ax.scatter(x,y, c=c)
ax.set_title("Title")
ax.set_xlabel("X label")
ax.set_label("Y label")
```

```python
data = [1,1,2,2,1,2,3,3,2,3,1,3,2]

fig, ax = plt.subplots()
ax.hist(data, bins = 4)
ax.set_title("Title")
ax.set_xlabel("X label")
ax.set_ylabel("Y label")
```

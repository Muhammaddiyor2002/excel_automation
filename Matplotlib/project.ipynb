```python
import matplotlib.pyplot as plt
%matplotlib inline 
```

```python
import csv
```

```python
with open("cars.csv", "r") as f:
    csv_reader = csv.DictReader(f)
    cars = list(csv_reader)
```

```python
cars[0]
```

```python
horse_power = [int(car['Horsepower']) for car in cars]
torque_power = [int(car['Torque']) for car in cars]
mpg = [int(car['Fuel City mpg']) for car in cars]
```

```python
fig, ax = plt.subplots()
ax.scatter(horse_power, torque_power, alpha=.25, c=mpg)
ax.set_xlabel("Horse_power")
ax.set_ylabel("Torque_power")
```

```python
import random
```

```python
def roll_dice():
    return random.randint(1, 6) + random.randint(1,6)
```

```python
roll_dice()
```

```python
data = [roll_dice() for _ in range(500)]
len(data)
```

```python
min(data)
```

```python
max(data)
```

```python
fig, ax = plt.subplots()
ax.hist(data, bins=11)
```
